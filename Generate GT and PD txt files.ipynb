```python
import os 
import glob
import json
import codecs
SET = "/set05-night"
DIR = "C:/Users/mhbt-/Desktop/gitStuffs/darkflow/kaistimages/output"+SET
```

```python
typeFolders = ['/lwir_005/result.json','/visible/result.json']
folder_dir = glob.glob(DIR+"/*")

try:
    folder_dir.remove(DIR+'\\predicted')
    folder_dir.remove(DIR+'\\predicted_005')
    folder_dir.remove(DIR+'\\ground-truth')
    folder_dir.remove(DIR+'\\results')
    folder_dir.remove(DIR+'\\main.py')
    folder_dir.remove(DIR+'\\results.py')
    # print(folder_dir)
except:
    pass


def readJson(directory):
    with open(directory, 'r') as content_file:
        content = content_file.read()
    startInd = 0
    endInd =0
    resList=[]
#     print(content)
    for i in range(len(content)):
        if(content[i]=='['):
            startInd = i
        elif(content[i]==']'):
            endInd =i+1
#             print(content[startInd:endInd])
            resList.append(json.loads(content[startInd:endInd]))
        
        
        
#     print(content)
    return resList


def getFileName(dir,delimit='.'):
    for i in list(range(len(dir))[::-1]):         
        if(dir[i] == delimit):break
    print(dir[:i])        
    return dir[:i]

def jsonDatatoMAPtxt(jsondata,bn,textSaveDir,predicted=False):
   
    for data in jsondata:
        txtfileName = bn+'_'+getFileName(data[0]['FileName'])
        resString=''
        for labelIndex in range(1,len(data)):
            label = data[labelIndex]['label']
            if(" " in label):
                label = getFileName(label,delimit=" ")
                
                
            left = data[labelIndex]['topleft']['x']
            top =  data[labelIndex]['topleft']['y']
            right =  data[labelIndex]['bottomright']['x']
            bottom = data[labelIndex]['bottomright']['y']
            conf = data[labelIndex]['confidence']
            
            if(not predicted):
                resString += label + " " + str(left) + " " + str(top)+ " " +str(right)+ " " +str(bottom)+ " " +"\n"
            else:
                resString += label + " "+str(conf) + " "+ str(left) + " " + str(top)+ " " +str(right)+ " " +str(bottom)+ " " +"\n"
                
 
        if(not predicted):
            file = open(textSaveDir+"/ground-truth/"+txtfileName+".txt","w") 
            file.write(resString) 
            file.close()
        else:
            file = open(textSaveDir+"/predicted_005/"+txtfileName+".txt","w") 
            file.write(resString) 
            file.close()
    
for directory in folder_dir:
    print(directory)
    baseFolderName = os.path.basename(directory)
    
    # Ground Truth | YOLO txt files
#     grdtruth_json_dir = directory+typeFolders[1]
#     grdtruth_json_data = readJson(grdtruth_json_dir)
#     #print(grdtruth_json_data)
#     jsonDatatoMAPtxt(grdtruth_json_data,baseFolderName,DIR,predicted=False)
    
     # Predicted
    pred_json_dir = directory+typeFolders[0]
    pred_json_data = readJson(pred_json_dir)
    #print(grdtruth_json_data)
    jsonDatatoMAPtxt(pred_json_data,baseFolderName,DIR,predicted=True)
    
    
    
    
            
        
        
    
    
```
